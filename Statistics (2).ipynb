 1.How do you calculate the mean, median, and mode of a dataset ?
To calculate the mean, median, and mode of a dataset, you just need a list of numbers and a few simple steps. Let’s break it down:

1. Mean (Average)
Steps:

Add all the numbers together.

Divide the total by how many numbers there are.

Formula:

Mean=∑𝑥𝑖𝑛
Where:

𝑥𝑖= each value in the dataset

𝑛= total number of values

Example: Dataset: 5, 10, 15 Mean = (5 + 10 + 15) ÷ 3 = 10

2. Median (Middle Value)
Steps:

Arrange the numbers in ascending order.

If the count is odd, the median is the middle number.

If the count is even, the median is the average of the two middle numbers.

Examples:

Odd: 3, 7, 9 → Median = 7

Even: 2, 4, 6, 8 → Median = (4 + 6) ÷ 2 = 5

3. Mode (Most Frequent Value)
Steps:

Count how many times each number appears.

The number(s) that appear most often is the mode.

Examples:

Dataset: 2, 4, 4, 6, 7 → Mode = 4

Dataset: 1, 2, 2, 3, 3 → Mode = 2 and 3 (bimodal)




```python
# 2. Write a Python program to compute the variance and standard deviation of a dataset ?

import math

# Sample dataset
data = [10, 15, 20, 25, 30]

# Calculate mean
mean = sum(data) / len(data)

# Calculate squared deviations
squared_diff = [(x - mean) ** 2 for x in data]

# Sample variance (divide by n-1)
variance = sum(squared_diff) / (len(data) - 1)

# Standard deviation = square root of variance
std_deviation = math.sqrt(variance)

# Output results
print(f"Data: {data}")
print(f"Mean: {mean}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_deviation}")

```

```python
# 9. Write a Python script to calculate covariance between two datasets ?
import numpy as np

# Example datasets
x = [10, 20, 30, 40, 50]
y = [12, 24, 33, 47, 55]

# 1️⃣ Manual calculation of covariance
def manual_covariance(x, y):
    n = len(x)
    mean_x = sum(x) / n
    mean_y = sum(y) / n
    covariance = sum((x[i] - mean_x) * (y[i] - mean_y) for i in range(n)) / (n - 1)
    return covariance

# 2️⃣ Covariance using NumPy
x_np = np.array(x)
y_np = np.array(y)
cov_matrix = np.cov(x_np, y_np, ddof=1)  # ddof=1 for sample covariance
cov_numpy = cov_matrix[0, 1]

# Print results
print("Manual Covariance:", manual_covariance(x, y))
print("NumPy Covariance:", cov_numpy)
```

```python
# 20.  Write a Python program to visualize covariance using a heatmap ?
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset: student scores in different subjects
data = {
    'Math': [88, 92, 80, 89, 100, 95, 78, 85],
    'Science': [90, 85, 88, 92, 95, 91, 76, 84],
    'English': [70, 75, 78, 72, 80, 77, 68, 74],
    'History': [65, 70, 72, 68, 75, 73, 60, 66]
}

# Create DataFrame
df = pd.DataFrame(data)

# Compute covariance matrix
cov_matrix = df.cov()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cov_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Covariance Matrix Heatmap")
plt.show()
```

```python
# 21. Use seaborn to create a correlation matrix for a dataset ?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Math': [88, 92, 80, 89, 100, 95, 78, 85],
    'Science': [90, 85, 88, 92, 95, 91, 76, 84],
    'English': [70, 75, 78, 72, 80, 77, 68, 74],
    'History': [65, 70, 72, 68, 75, 73, 60, 66]
}

# Create DataFrame
df = pd.DataFrame(data)

# Compute correlation matrix
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# 22. Generate a dataset and implement both variance and standard deviation computations?
import numpy as np
import pandas as pd

# Generate a dataset of 1000 values from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=75, scale=12, size=1000)  # mean=75, std=12

# Convert to pandas Series
df = pd.Series(data)

# Compute variance and standard deviation
variance = df.var()  # sample variance (ddof=1)
std_dev = df.std()   # sample standard deviation (ddof=1)

# Display results
print("Descriptive Statistics:")
print(f"Mean: {df.mean():.2f}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
```

```python
# 23. Visualize skewness and kurtosis using Python libraries like matplotlib or seaborn?
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# Simulate a dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

# Convert to pandas Series
df = pd.Series(data)

# Calculate skewness and kurtosis
skew_val = skew(df)
kurt_val = kurtosis(df)  # excess kurtosis

# Plot histogram with KDE
plt.figure(figsize=(8, 5))
sns.histplot(df, bins=30, kde=True, color='skyblue')
plt.title(f"Histogram with KDE\nSkewness = {skew_val:.2f}, Kurtosis = {kurt_val:.2f}")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# 24.  Implement the Pearson and Spearman correlation coefficients for a dataset.
import pandas as pd

# Create a sample dataset
data = {
    'Math': [88, 92, 80, 89, 100, 95, 78, 85],
    'Science': [90, 85, 88, 92, 95, 91, 76, 84],
    'English': [70, 75, 78, 72, 80, 77, 68, 74]
}

df = pd.DataFrame(data)
from scipy.stats import pearsonr

# Example: Pearson correlation between Math and Science
pearson_corr, p_value_pearson = pearsonr(df['Math'], df['Science'])
print(f"Pearson Correlation (Math vs Science): {pearson_corr:.2f}, p-value: {p_value_pearson:.4f}")
```
